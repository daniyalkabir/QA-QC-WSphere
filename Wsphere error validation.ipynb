```python
# Sites Tab

# sites needs to be in correct order as described in the meta data
# site name does not need checking it is a text value
# site type, sample matrix type, institution type need to be checked since they have to be a fixed option
# admin0,1,2,3, address line, postal code does not need to be checked since they are text values
# address line and postal code don't need to be checked
# latitude and longitude need to be checked
# geodata link and contact person don't need to be checked since they are text values
# phone should NOT be checked to make sure that all values are unhyphenated
# email should NOT be checked to make sure it is in the correct format
# sample collection method needs to be tested based on the sample method options
# Population Served and Annual Average Daily Flow do not need to be tested
# flow units needs to be tested
# sampling frequency an d active needs not to be tested based on options provided in dictionary



# In this code, pd.read_excel() reads the Excel file named "example_file.xlsx" and reads the tab named "sites". 
# The function returns a DataFrame object.

# We then use the str.match() method with a regular expression pattern r'^[A-Z]{3}_[A-Z]{3}_Site_\d+$' 
# to check if all values in the "Site ID" column match the format COUNTRY_CODE_REGION_Site_NUMBER. The 
# regular expression pattern matches any string that starts with three uppercase letters, followed by an 
# underscore, followed by another three uppercase letters, followed by "Site", and ends with one or more digits.

# If all values in the "Site ID" column match the pattern, the code prints a message saying that 
# the column is in the correct format. If not, the code prints a message saying that the column is 
# not in the correct format.





import pandas as pd



# Read the Excel file with the "sites" tab
excel_file = pd.read_excel('/Users/daniyalkabir/Downloads/Austria_WSPHERE_Data_Collection_Form_v7_2023_02_09_dd_jb.xlsx', sheet_name='2. Sites')



#SITE ID 

# Check if the "Site ID" column is in the correct format
mask = excel_file['Site ID'].str.match(r'^[A-Z]{3}_[A-Z]{3}_Site_\d+$')
if mask.all():
    print("The 'Site ID' column is in the correct format.")
else:
    print("The 'Site ID' column is not in the correct format.")
    print("Incorrect values:")
    print(excel_file.loc[~mask, 'Site ID'])


#SITE TYPE
excel_file['Site Type'] = excel_file['Site Type'].apply(str.strip)

# Define a list of allowed site types
allowed_site_types = ['sewer', 'pump station in sewer', 'open sewer', 'sewage lagoon', 'wastewater treatment plant',
                      'influent at treatment plant', 'effluent at treatment plant', 'primary sludge', 'septic tank', 
                      'river', 'other']

# Check if the "Site Type" column contains only allowed values
mask = excel_file['Site Type'].isin(allowed_site_types)
if mask.all():
    print("The 'Site Type' column contains only allowed values.")
else:
    print("The 'Site Type' column contains some disallowed values.")
    print("Disallowed values:")
    print(excel_file.loc[~mask, 'Site Type'])
    
    

#SAMPLE MATRIX TYPE
excel_file['Site Type'] = excel_file['Site Type'].apply(str.strip)
excel_file['Sample Matrix Type'] = excel_file['Sample Matrix Type'].apply(str.strip)

# Define a list of allowed sample matrix types
allowed_matrix_types = ['wastewater', 'fecal sludge', 'septage', 'river', 'other']

# Check if the "Sample Matrix Type" column contains only allowed values
mask = excel_file['Sample Matrix Type'].isin(allowed_matrix_types)
if mask.all():
    print("The 'Sample Matrix Type' column contains only allowed values.")
else:
    print("The 'Sample Matrix Type' column contains some disallowed values.")
    print("Disallowed values:")
    print(excel_file.loc[~mask, 'Sample Matrix Type'])
    

# INSTITUTION TYPE
excel_file['Institution Type'] = excel_file['Institution Type'].apply(str.strip)

# Define a list of allowed institution types
allowed_institution_types = ['correctional institution', 'hospital', 'residential building', 'long term care', 'recreational area', 'hotel', 'commercial building/complex', 'educational dorm', 'educational other', 'social services building', 'ship', 'airplane', 'other']

# Check if the "Institution Type" column contains only allowed values
mask = excel_file['Institution Type'].isin(allowed_institution_types)
if mask.all():
    print("The 'Institution Type' column contains only allowed values.")
else:
    print("The 'Institution Type' column contains some disallowed values.")
    print("Disallowed values:")
    print(excel_file.loc[~mask, 'Institution Type'])

# LATITUDE AND LONGITUDE

excel_file['Latitude'] = excel_file['Latitude']
excel_file['Longitude'] = excel_file['Longitude']

# Check if "Latitude" and "Longitude" columns have valid values
invalid_latitudes = excel_file[~excel_file['Latitude'].between(-90, 90)]
invalid_longitudes = excel_file[~excel_file['Longitude'].between(-180, 180)]

# Print the results
if invalid_latitudes.empty and invalid_longitudes.empty:
    print("All latitude and longitude values are valid.")
elif invalid_latitudes.empty:
    print("All latitude values are valid.")
elif invalid_longitudes.empty:
    print("All ongitude values are valid.")
else:
    if not invalid_latitudes.empty:
        print("Invalid latitude values:")
        print(invalid_latitudes[['Site ID', 'Latitude']])
    if not invalid_longitudes.empty:
        print("Invalid longitude values:")
        print(invalid_longitudes[['Site ID', 'Longitude']])

#FLOW UNITS
excel_file['Flow Units'] = excel_file['Flow Units'].apply(str.strip)

# Define the expected flow units
expected_flow_units = ['m^3/day', 'MGD']

# Check if "FLOW UNITS" column has expected values
invalid_flow_units = excel_file[~excel_file['Flow Units'].isin(expected_flow_units)]

# Print the results
if invalid_flow_units.empty:
    print("All flow units are valid.")
else:
    print("Invalid flow units:")
    print(invalid_flow_units[['Site ID', 'Flow Units']])

```

```python
# SAMPLE ID TAB

import pandas as pd
import re

# Load the data file into a Pandas DataFrame
data_column = pd.read_excel('/Users/daniyalkabir/Downloads/Austria_WSPHERE_Data_Collection_Form_v7_2023_02_09_dd_jb.xlsx', sheet_name='4. Samples')



data = data_column["Sample ID"]



# Define regular expression patterns to match the values
pattern1 = r'^AUT_CAR_Site_1_S\d+$'
pattern2 = r'^AUT_CAR_Site_2_S\d+$'
pattern3 = r'^AUT_VOR_Site_1_S\d+$'
pattern4 = r'^AUT_VOR_Site_2_S\d+$'
pattern5 = r'^AUT_SAL_Site_1_S\d+$'
pattern6 = r'^AUT_SAL_Site_2_S\d+$'


check = True
# Iterate over the values in the column and check if they match either pattern
for value in data:
    if not re.match(pattern1, value) and not re.match(pattern2, value) and not re.match(pattern3, value) and not re.match(pattern4, value) and not re.match(pattern5, value) and not re.match(pattern6, value):
        print(f"Invalid format for value: {value}")
        check = False

if check:
    print("No error in Sample ID")
```
